**1. Using Pandas and Matplotlib, recreate the following donut chart to the best of your ability.**

Levy Fig2C.jpg

```python
# Get the necessary libraries below
import pandas  as pd
import matplotlib.pyplot as plt
```

```python
# Use the following dataframe to create the donut chart

figure_one_df = pd.DataFrame({
    'City':{0:'Paris, France',
            1:'Washington, D.C., United States',
            2:'Geneva, Switzerland',
            3:'New York, United States',
            4:'The Hague, Netherlands',
            5:'Vancouver, Canada'},
    'Count':{
            0:8,
            1:7,
            2:2,
            3:1,
            4:1,
            5:1
    }
})
```

```python
# Write the chart creation code below
plt.figure(figsize=(10,8))


wedges, texts, autotexts = plt.pie(
    figure_one_df['Count'],
    labels=None,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'width': 0.4},
    pctdistance=0.75
)


centre_circle = plt.Circle((0,0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.legend(
    wedges,
    figure_one_df['City'],
    title='Cities',
    loc='center left',
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title('Arbitral Seats in Levy Arbitrations', fontsize=16)
plt.axis('equal')

plt.show()
```
